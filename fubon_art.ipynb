```python
import os
import re
import csv
import requests as req
from bs4 import BeautifulSoup as bs
from urllib.parse import urljoin
from requests.utils import requote_uri

base_url = "https://www.fubonartmuseum.org"
exhs_url = "https://www.fubonartmuseum.org/Exhibitions"
museum_name = "富邦美術館"

resp = req.get(exhs_url)
html = bs(resp.text, "html.parser")
exhs = html.find_all("a", class_="fb-exhibitions-card")

rows = []  # 用來存所有展覽資料

for exh in exhs:
  ex_link = exh["href"]
  link = urljoin(base_url, ex_link)

  info_group = exh.find_all("div", class_="info_group")

  # 展覽名稱
  title = ""
  try:
      ex_title = info_group[0].find("h2", class_="font-h2 font-bold")
      title = ex_title.text.strip()
  except Exception:
      pass


  date_place_group = info_group[2].find_all("p")
  # 展覽日期
  ex_date = ""
  try:
      ex_date = date_place_group[0].text.strip()
  except Exception:
      pass

  # 展覽地點
  place = ""
  try:
      place = date_place_group[1].text.strip()
  except Exception:
      pass

  # 展覽圖片
  img = ""
  try:
      ex_img = exh.find("img")["src"]
      img = requote_uri(ex_img)
  except Exception:
      pass

  rows.append([museum_name, title, link, img, ex_date, place])
  print(link, title, ex_date, place, img)

# --- 將結果輸出為 CSV 檔 ---
output_file = "fubon_artmuseum_exhibitions.csv"

with open(output_file, "w", newline="", encoding="utf-8-sig") as f:
    writer = csv.writer(f)
    # 寫入欄位名稱
    writer.writerow(["館別", "展覽名稱", "展覽日期", "展覽地點", "展覽連結", "展覽圖片"])
    # 寫入內容
    for row in rows:
        writer.writerow(row)

print(f"✅ 已輸出 {len(rows)} 筆資料至 {output_file}")
```
